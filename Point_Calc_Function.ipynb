# Function for Calculating Team's Weight

```python
def Points_cal(df):
    df = df.sort_values(by='Matches',ascending=False).reset_index(drop=True)
    df = df.iloc[:11,:]    # Selecting Top 11 players
    x = df.iloc[:,3:]      # Independent variables
    y = df['pts']          # Dependent variables
    clf = linear_model.LinearRegression()
    clf.fit(x,y)
    val = list(y.values)
    Team_weight = 0
    for i in val:
       Team_weight += i
    print(Team_weight/df['Matches'].values[0])
```

```python
import tensorflow as tf
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense
from warnings import filterwarnings
filterwarnings("ignore")
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'
import keras
from keras.utils import to_categorical, plot_model
from keras import losses
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.metrics import accuracy_score, log_loss, recall_score
import matplotlib.pyplot as plt
```

```python
# Creating model
model = Sequential()
model.add(Dense(100, activation="relu", input_dim=83, kernel_regularizer=keras.regularizers.l2(l=0.001)))
model.add(Dense(75, activation="relu", kernel_regularizer=keras.regularizers.l2(l=0.001)))
model.add(Dense(50, activation="relu", kernel_regularizer=keras.regularizers.l2(l=0.001)))
model.add(Dense(35, activation="relu", kernel_regularizer=keras.regularizers.l2(l=0.001)))
model.add(Dense(13, activation="sigmoid"))
model.compile(optimizer=keras.optimizers.Adam(lr=0.0001), loss='categorical_crossentropy', metrics=["accuracy"])
```

```python
history = model.fit(X_train,y_train,epochs=1000, validation_data=(X_test,y_test), batch_size=25)
```

```python
pred = model.predict(X_test)
e = model.evaluate(X_test,y_test)
print ("Loss: ", e[0])
print ("Accuracy on the test set:"" ", e[1])
```
